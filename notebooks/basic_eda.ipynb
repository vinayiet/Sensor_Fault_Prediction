```python
def get_cols_with_zero_std(df: pd.DataFrame):
    cols_to_drop = []
    nums_cols = [col for col in df.columns if df[col].dtype != 'O']
    for col in nums_cols:
        if df[col].std() == 0:
            cols_to_drop.append(col)
    return cols_to_drop

def get_redundant_cols(df: pd.DataFrame, missing_thresh=.7):
    cols_mssing_ratios = df.isna().sum().div(df.shape[0])
    cols_to_drop = list(cols_mssing_ratios[cols_mssing_ratios > missing_thresh].index)
    return cols_to_drop
```

```python
cols_to_drop = get_redundant_cols(wafers, missing_thresh=.7)
cols_to_drop
```

```python
cols_to_drop2 = get_cols_with_zero_std(df=wafers)
cols_to_drop2.append('wafer')
cols_to_drop2
```

```python
cols_to_drop_final = cols_to_drop + cols_to_drop2
```

```python
X, y = wafers.drop(cols_to_drop_final, axis=1), wafers['Good/Bad']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler

imputer = KNNImputer(n_neighbors=3)
processing_pipeline = Pipeline(
    steps=[('Imputer', imputer), ('Scaler', RobustScaler())]
)

processing_pipeline
```

```python
X_trans = processing_pipeline.fit_transform(X)
print("Shape of transformed features set: ", X_trans.shape)
X_trans
```
